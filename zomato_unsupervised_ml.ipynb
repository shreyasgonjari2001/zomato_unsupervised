## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# merging two dataset on common key "Name" and "Restaurant"
dataset = pd.merge(restaurant_data, restaurant_reviews, left_on="Name", right_on= "Restaurant", how="inner")
````

````python
# function for transforming cast column into float
def to_float(x):
  return float(x.replace(",", ""))
````

````python
dataset["Cost"] = dataset["Cost"].apply(to_float)
````

````python
dataset["Rating"].value_counts().plot(kind="bar")  # Rating has "like" value which prevents us from converting it to float
````

````python
# replacing Like with 0
dataset["Rating"][dataset["Rating"] == "Like"] = 0
````

````python
# now casting Rating column to float datatype for analysis purpose
dataset["Rating"] = dataset["Rating"].astype(float)
````

````python
# function of extracting the metadata
def get_reviews(x):
  try:
    l = x.strip().split(" , ")
  except AttributeError:
    l = str(x).split(" , ")
  for i in l:
    if "Reviews" in i:
      return int(i.split(" ")[0])


def get_followers(x):
  try:
    l = x.strip().split(" , ")
  except AttributeError:
    l = str(x).split(" , ")
  for i in l:
    if "Followers" in i:
      return int(i.split(" ")[0])
````

### 3. Categorical Encoding

````python
# Encode your categorical columns
# dropping less usefull features like names, links etc.
dataset.drop(["Links"], axis=1, inplace=True)
````

````python
# building categorical encoder for the feature "Cuisines"
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.utils.validation import check_is_fitted

class CategoricalEncoder(BaseEstimator, TransformerMixin):
  def __init__(self):
    self.features_out_ = None
  def fit(self, X, y=None):
    self.features_out_ = {}
    j = 0
    for i in range(X.shape[0]):
      cuisines = X.iloc[i][0].split(", ")
      for cuisine in cuisines:
        if cuisine not in self.features_out_:
          self.features_out_[cuisine] = j
          j += 1
    self.n_features_out_ = len(self.features_out_)
    return self

  def transform(self, X):
    check_is_fitted(self)
    assert X.shape[1] == 1
    X_out = np.zeros((X.shape[0], self.n_features_out_))
    for i in range(X.shape[0]):
      cuisines = X.iloc[i][0].split(", ")
      for cuisine in cuisines:
        X_out[i, self.features_out_[cuisine]] = 1
    return X_out
````

````python
cat_encoder = CategoricalEncoder()
cat_encoder.fit(dataset[["Cuisines"]])
````

````python
cat_data = cat_encoder.transform(dataset[["Cuisines"]])
````

#### 1. Expand Contraction

````python
import string
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


nltk.download('punkt')  # Download the 'punkt' resource for sentence tokenization
nltk.download('stopwords')
nltk.download('wordnet')

dataset['Review'] = dataset['Review'].astype(str)
# Define a function to clean text
def clean_text(text):
    # Convert text to lowercase
    text = text.lower()
    # Remove punctuation
    text = text.translate(str.maketrans('', '', string.punctuation))
    # Remove numbers
    text = ''.join([i for i in text if not i.isdigit()])
    # Tokenize text
    tokens = word_tokenize(text)
    # Remove stopwords
    tokens = [word for word in tokens if word not in stopwords.words('english')]
    # Lemmatize tokens
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(tokens)


# Apply the cleaning function to the reviews
dataset['cleaned_review'] = dataset['Review'].apply(clean_text)
````
